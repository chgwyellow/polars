# Select Rows - Indexing with `[]`

```python
import polars as pl
```

```python
csv_file = "data/titanic.csv"
```

```python
df = pl.read_csv(csv_file)
df.head(3)
```

## Select Individual Row

```python
df[0] # first row
```

## Select Multiple Rows

```python
df[[2, 3]]
```

## Slice

```python
df[:2]
```

## Range

```python
df[range(2, 3)]
```

## Numpy Array

```python
import numpy as np

df[np.arange(0, 3)]
```

## Data types are not accepted in []

We can not pass a `list` of **Boolean values** in `[]`

```python
df[[True for _ in range(len(df))]]
```

Neither is the `Series`.

```python
df[df["Age"] > 30]
```
