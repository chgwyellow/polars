# Select Rows - with Filter and Expression API

```python
import polars as pl
```

```python
csv_file = "data/titanic.csv"
```

```python
df = pl.read_csv(csv_file)
df.head(3)
```

## Selecting rows with a condition using `filter`

```python
df.filter(
    pl.col("Pclass") == 1 # Put conditions here
).select(
    "PassengerId", "Pclass"
).head(3)
```

### Mathematical Operator

```python
df.filter(
    pl.col("Parch") > 1
).select(
    "PassengerId", "Parch", "SibSp"
).head(3)
```

### Text Operator

```python
df.filter(
    pl.col("Parch").gt(pl.col("SibSp")) # gt stands for greater than
).select(
    "PassengerId", "Parch", "SibSp"
).head(3)
```

### Keyword Filter

```python
df.filter(
    Parch = 3
).select(
    "PassengerId", "Parch", "SibSp"
).head(3)
```

## Filtering on a Boolean List

```python
is_first_class = [True if fc == 1 else False for fc in df["Pclass"]]
is_first_class[:3]
```

```python
df.filter(
    is_first_class
).head(5)
```

## Filtering on a Boolean Column

```python
df.with_columns(
    less_than_30_boolean = pl.col("Age") < 30
).filter(
    pl.col("less_than_30_boolean")
).select(
    "Pclass", "Name", "Age"
).head(4)
```

## Negation of a Condition

```python
df.with_columns(
    less_than_30_boolean = pl.col("Age") < 30
).filter(
    ~pl.col("less_than_30_boolean") # ~ is not in 
).select(
    "Pclass", "Name", "Age"
).head(4)
```

## Partitioning a `DataFrame`

```python
df_pclass_dict = (
    df.partition_by(by="Pclass", as_dict=True) # partition_by uses tuple to be the key
)
```

```python
df_pclass_dict.keys()
```

```python
df_pclass_dict[(1,)].head(2)
```

## Filter in Lazy Mode

```python
pl.scan_csv(
    csv_file
).filter(
    pl.col("Age") > 30
).select(
    "Pclass", "Name", "Age"
)
```

`.explain()` show the optimized plan, and `filter` operation is behind scanning CSV.

```python
print(pl.scan_csv(
    csv_file
).filter(
    pl.col("Age") > 30
).select(
    "Pclass", "Name", "Age"
).explain())
```

```python
print(pl.scan_csv(
    csv_file
).filter(
    pl.col("Age") > 30
).select(
    "Pclass", "Name", "Age"
).explain(engine="streaming"))
```
